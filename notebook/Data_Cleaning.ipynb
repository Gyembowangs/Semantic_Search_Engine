```python
# Drop the missing values from product_name, product_category_tree,description
df = df.dropna(subset=[
    'product_name',
    'product_category_tree',
    'description',
    'uniq_id',
    'image',
])
```

```python
df.loc[:, 'retail_price'] = df['retail_price'].fillna(df['retail_price'].median())
df.loc[:, 'discounted_price'] = df['discounted_price'].fillna(df['discounted_price'].median())
df.loc[:, 'brand'] = df['brand'].fillna('Unknown')
df
```

## 7. Text Cleaning and Feature Engineering

```python
# cleaning the text function or text normalization
def clean_text(text):
    text = text.lower()                      # lowercase
    text = re.sub(r'<.*?>', '', text)        # remove HTML tags
    text = re.sub(r'[^a-z0-9 ]', ' ', text)  # remove special characters
    text = re.sub(r'\s+', ' ', text).strip() # remove extra spaces
    return text
```

### Apply Text Cleaning to Relevant Columns

```python
# Apply cleaning
df=df.copy()
df.loc[:, 'product_name'] = df['product_name'].fillna('').apply(clean_text)
df.loc[:, 'description'] = df['description'].fillna('').apply(clean_text)
df.loc[:, 'brand'] = df['brand'].fillna('').apply(clean_text)

# Extract category BEFORE cleaning it
df.loc[:, 'category'] = df['product_category_tree'] = df['product_category_tree'].str.replace(">>", " ", regex=False)
df.loc[:, 'category'] = df['category'].fillna('').apply(clean_text)
df

# We clean the text before combining it so that the final
#  product_text is already clean and consistent.
```

### Drop Original Product Category Tree Column

```python
df.drop('product_category_tree', axis=1, inplace=True)
df
```

### Combine Text Columns into a Single Feature

```python
df.loc[:, 'product_text'] = (
    df['product_name'].fillna('') + " " +
    df['description'].fillna('') + " " +
    df['category'].fillna('') + " " +
    df['brand'].fillna('')
)

df.loc[:, 'product_text'] = df['product_text'].apply(clean_text)
df
```
